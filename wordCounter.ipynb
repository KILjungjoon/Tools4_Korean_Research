# 1.일반적인 wordCounter

```python
def wordsCount(str) :
  words = []
  words = text.split()
  wfreq=[words.count(w) for w in words]
  result=dict(zip(words,wfreq))
  return result
```

```python
text=open('/content/소나기_tag.txt', 'r', encoding='utf-8-sig').read()
text=text.replace('+', ' ')
text=text.replace('__', '')
text
```

```python
print(wordsCount(text))
```

# 2.내림차순 wordCounter

```python
from collections import Counter

word_count = Counter()

word_count.update((word for word in text.split()))

for word, count in word_count.most_common():
    print(word, count)
```
